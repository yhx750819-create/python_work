 int() 和 float() 转换

```python
# 1. 基本转换
print(int(3.14))      # 3（直接截断，不是四舍五入！）
print(float(3))       # 3.0
print(int("123"))     # 123
print(float("123.45")) # 123.45

# 2. 常见错误
try:
    print(int("123.45"))   # ValueError: 字符串包含小数点
except ValueError as e:
    print(f"错误：{e}")

try:
    print(int("abc"))      # ValueError: 非数字字符串
except ValueError as e:
    print(f"错误：{e}")

# 3. 安全转换函数（实际开发常用）
def safe_int(value, default=0):
    """安全转换为整数，失败返回默认值"""
    try:
        return int(value)
    except (ValueError, TypeError):
        return default

def safe_float(value, default=0.0):
    """安全转换为浮点数"""
    try:
        return float(value)
    except (ValueError, TypeError):
        return default

# 测试安全转换
print(safe_int("123"))        # 123
print(safe_int("123.45"))     # 0（默认值）
print(safe_int("abc", 100))   # 100（自定义默认值）
print(safe_float("123.45"))   # 123.45
print(safe_float("abc"))      # 0.0

# 4. 实际应用场景
# 场景1：用户输入处理
user_input = input("请输入年龄：")
age = safe_int(user_input, 0)  # 避免程序崩溃
if age <= 0:
    print("请输入有效的年龄！")

# 场景2：数据清洗（信管专业重要！）
data = ["123", "45.6", "invalid", "78.9", "100"]
clean_data = []
for item in data:
    # 尝试转换为浮点数，成功则保留
    value = safe_float(item)
    if value != 0.0 or item == "0":  # 0也可能是有效数据
        clean_data.append(value)

print(f"清洗后数据：{clean_data}")
# 输出：[123.0, 45.6, 78.9, 100.0]

# 5. 类型转换链
value = "123.45"
# 字符串 → 浮点数 → 整数
value_as_int = int(float(value))  # 先转浮点再转整数
print(f"'{value}' → float → int = {value_as_int}")  # 123

# 6. 布尔值转换的特殊性
print(f"int(True) = {int(True)}")      # 1
print(f"int(False) = {int(False)}")    # 0
print(f"float(True) = {float(True)}")  # 1.0

# 这在条件判断中很有用
result = 5 * True    # 5 × 1 = 5
result2 = 5 * False  # 5 × 0 = 0

# 7. 进制转换（仅int()支持）
print(int("1010", 2))     # 二进制 → 10
print(int("FF", 16))      # 十六进制 → 255
print(int("0o77", 8))     # 八进制 → 63
```

```python
#程序依次询问周一至周五每天的出勤人数
attendance = []
days = ["周一", "周二", "周三", "周四", "周五"]
for day in days:
    num = input(f"请输入{day}的出勤人数：")
    attendance.append(int(num))
#计算一周的总出勤人数
total_attendance = sum(attendance)
print(f"本周总出勤人数为：{total_attendance}人")
#平均每日出勤人数
average_attendance = total_attendance / len(days)
print(f"平均每日出勤人数为：{average_attendance:.2f}人")
#找出出勤人数最多和最少的日子及其人数   
max_attendance = max(attendance)
min_attendance = min(attendance)    
max_day = days[attendance.index(max_attendance)]
min_day = days[attendance.index(min_attendance)]
print(f"出勤人数最多的是{max_day}，人数为{max_attendance}人")
print(f"出勤人数最少的是{min_day}，人数为{min_attendance}人")
```
